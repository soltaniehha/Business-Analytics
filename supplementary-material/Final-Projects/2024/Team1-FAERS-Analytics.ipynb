## 6.1. Adverse Reaction Frequecies Analysis

**Question: What are the most common adverse reactions reported?**


In this analysis, we are working on summarizing the frequencies of initial FAERS cases by specific medical events, namely reactions, from both 2023 Q3 and 2023 Q4 data, to figure out what reactions are the most popular. To futhur drill down into the data, we want to figure out in each medical event, who are the age groups that are reported the most.

```python
# join the demo table with the reaction table in this case with looking at the initial case only
joined_df1 = demo_append[demo_append['i_f_code']=='I'].merge(reac_append , how='inner', on='primaryid')
# Grouping by 'key_column' and summing 'value1' and 'value2'
grouped_df = joined_df1.groupby(['pt', 'age_grp', 'sex']).size().reset_index(name='counts')
sorted_group_sizes1 = grouped_df.sort_values(by='counts', ascending=False).reset_index()
```

```python
# import the plot lab for plotting
import matplotlib.pyplot as plt
grouped_df_one = joined_df1.groupby(['pt']).size().reset_index(name='counts')
sorted_group_sizes_one_group = grouped_df_one.sort_values(by='counts', ascending=False).reset_index()

plot_data = sorted_group_sizes_one_group.head(10)
plt.figure(figsize=(10, 10))  # Adjust the figure size as necessary
bars = plt.bar(plot_data['pt'], plot_data['counts'], color='skyblue')

# Adding annotations
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{int(bar.get_height()):,}', ha='center', va='bottom', fontsize=8)

plt.xlabel('Medical Event')
plt.ylabel('Cases')
plt.title('Number of FAERS Cases per Medical Event')
plt.xticks(rotation=60)  # Rotate labels to make them readable if necessary
plt.show()
```

```python
# drill down the case numbers per medical event by age group
import numpy as np

select_pt = plot_data.pt.unique()
plot_data_three_group = sorted_group_sizes1[sorted_group_sizes1['pt'].isin(select_pt)]

def age_mapping(age):
    if age == 'A':
     return 'Adult'
    elif age == 'C':
     return 'Child'
    elif age == 'T':
     return 'Adolescent'
    elif age == 'I':
     return 'Infant'
    elif age == 'E':
     return 'Elderly'
    elif age == 'N':
     return 'Neonate'

plot_data_three_group['age']= plot_data_three_group['age_grp'].map(age_mapping)
#print(plot_data_three_group)
```

```python
pivot_df_raw = plot_data_three_group.pivot_table(index='pt', columns=['age'], values='counts', aggfunc='sum')
pivot_df_raw['row_total']=pivot_df_raw['Adolescent']+pivot_df_raw['Adult']+pivot_df_raw['Child']+pivot_df_raw['Elderly']+pivot_df_raw['Infant']+pivot_df_raw['Neonate']
pivot_df = pivot_df_raw.sort_values(by='row_total',ascending = False)
#print(pivot_df)
sorted_pivot_df = pivot_df.iloc[:,:6]
#print(sorted_pivot_df)
```

```python
# Plotting
sorted_pivot_df.plot(kind='bar', figsize=(12, 8))
plt.xlabel('Medical Event')
plt.ylabel('Cases')
plt.title('Number of FAERS Cases per Medical Event by age group')
plt.xticks(rotation=40)
plt.legend(title='age_groups')
plt.tight_layout()
plt.show()
```

## 6.2. Comparative Outcomes by Report Source Analysis

**Question: How do the sources of FDA adverse event reports influence the reported outcomes within a given timeframe?**

This analysis explores the relationship between the reporter demographic—be it consumers, health professionals, or others—and the severity of reported outcomes to inform healthcare stakeholders and enhance patient safety measures.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

rpsr_q3 = data['RPSR23Q3']
outc_q3 = data['OUTC23Q3']
rpsr_q4 = data['RPSR23Q4']
outc_q4 = data['OUTC23Q4']

rpsr_combined = pd.concat([rpsr_q3, rpsr_q4], ignore_index=True)
outc_combined = pd.concat([outc_q3, outc_q4], ignore_index=True)

rpsr_combined.columns = rpsr_combined.columns.str.upper()
outc_combined.columns = outc_combined.columns.str.upper()


# Mapping of report source codes to their full descriptions
report_source_definitions = {
    'FGN': 'Foreign',
    'SDY': 'Study',
    'LIT': 'Literature',
    'CSM': 'Consumer',
    'HP': 'Health Professional',
    'UF': 'User Facility',
    'CR': 'Company Representative',
    'DT': 'Distributor',
    'OTH': 'Other'
}

# Mapping of outcome codes to their full descriptions
outcome_code_definitions = {
    'DE': 'Death',
    'LT': 'Life-Threatening',
    'HO': 'Hospitalization - Initial or Prolonged',
    'DS': 'Disability',
    'CA': 'Congenital Anomaly',
    'RI': 'Required Intervention to Prevent Permanent Impairment/Damage',
    'OT': 'Other Serious (Important Medical Event)'
}

df_rpsr_outc_combined = pd.merge(rpsr_combined, outc_combined, on='PRIMARYID', how='inner')


df_rpsr_outc_combined['RPSR_COD'] = df_rpsr_outc_combined['RPSR_COD'].map(report_source_definitions)
df_rpsr_outc_combined['OUTC_COD'] = df_rpsr_outc_combined['OUTC_COD'].map(outcome_code_definitions)


outcomes_by_source_combined = df_rpsr_outc_combined.groupby('RPSR_COD')['OUTC_COD'].value_counts().unstack().fillna(0)
cubehelix_palette = sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)

plt.figure(figsize=(12, 8))
outcomes_by_source_combined.plot(kind='bar', stacked=True, colormap=cubehelix_palette)
plt.title('Outcomes by Report Source for Q3 and Q4')
plt.xlabel('Report Source')
plt.ylabel('Count of Outcomes')
plt.legend(title='Outcome Code', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.subplots_adjust(left=0.1, right=0.75, top=0.9, bottom=0.1)
plt.xticks(rotation=45)
plt.show()
```

**Summary:**


1. **Distribution of Reports by Source:** Most of the adverse event reports seem to come from health professionals, followed by consumers. There are relatively few reports from foreign sources or literature.

2. **Report Outcomes:** The majority of the adverse event reports submitted by consumers and health professionals resulted in hospitalization, which is depicted by the dominant section in their respective bars.

3. **Serious Events:** Life-threatening outcomes and those requiring intervention to prevent permanent impairment/damage appear to be a significant portion of the reports, especially those submitted by health professionals.


4. **Disability and Death:** There's a noticeable number of reports associated with disability and death, predominantly in reports submitted by health professionals. This might suggest that health professionals report more severe cases.

5. **Less Common Outcomes:** Congenital anomaly and other serious important medical events are less commonly reported outcomes across all report sources.


## 6.3. Analysis of Therapy Duration and Off-Label Drug Usage Trends

**Question: How do therapy durations impact the frequency of adverse reactions?**

In this section, we explore the correlations between the length of therapy and the incidence of adverse reactions, as well as investigate the extent of off-label drug usage.

**Data Cleaning and Preprocessing:**

**Date Conversion and Duration Calculation:**
To work with therapy durations, we need to convert various date formats into a uniform datetime object and calculate the duration in days based on different units.

```python
def convert_to_datetime(date_str):
    if pd.isna(date_str):  # Handle NaN values
        return None

    # Check if the input is already a datetime object
    if isinstance(date_str, pd.Timestamp):
        return date_str

    # Convert to string for consistent handling
    date_str = str(int(date_str))  # Convert float to int, then to string

    if len(date_str) == 8:  # YYYYMMDD
        try:
            return pd.to_datetime(date_str, format='%Y%m%d')
        except ValueError:
            return None
    elif len(date_str) == 6:  # YYYYMM
        try:
            return pd.to_datetime(date_str, format='%Y%m')
        except ValueError:
            return None
    elif len(date_str) == 4:  # YYYY
        try:
            return pd.to_datetime(date_str, format='%Y')
        except ValueError:
            return None
    else:
        # Handle invalid or unexpected formats
        return None
```

```python
def convert_to_days(row):
    dur = row['dur']
    dur_cod = row['dur_cod']

    if dur_cod == 'YR':
        return dur * 365  # Assuming a year has 365 days
    elif dur_cod == 'MON':
        return dur * 30  # Assuming a month has 30 days
    elif dur_cod == 'WK':
        return dur * 7
    elif dur_cod == 'DAY':
        return dur
    elif dur_cod == 'HR':
        return dur / 24  # Convert hours to days
    elif dur_cod == 'MIN':
        return dur / (24 * 60)  # Convert minutes to days
    elif dur_cod == 'SEC':
        return dur / (24 * 60 * 60)  # Convert seconds to days
    else:
        return None  # Handle unknown dur_cod values
```

**Preprocessing Application:**
We apply the previous functions to our therapy data, cleaning and structuring it for analysis.

```python
def preprocess_data(df):
    df['start_dt'] = df['start_dt'].apply(convert_to_datetime)
    df['end_dt'] = df['end_dt'].apply(convert_to_datetime)

    # Fill null values in dur and dur_cod if both start and end dates are present
    df.loc[df['start_dt'].notnull() & df['end_dt'].notnull() & df['dur'].isnull() & df['dur_cod'].isnull(), 'dur'] = (df['end_dt'] - df['start_dt']).dt.total_seconds() / (24 * 3600)
    df.loc[df['start_dt'].notnull() & df['end_dt'].notnull() & df['dur'].isnull() & df['dur_cod'].isnull(), 'dur_cod'] = 'DAY'

    # Calculate duration in seconds and store in a new column
    df['dur_days'] = df.apply(convert_to_days, axis=1)

    # # Fill missing start or end dates when duration is available - broken
    # df.loc[df['start_dt'].isnull() & ~df['end_dt'].isnull() & ~df['dur'].isnull() & ~df['dur_cod'].isnull(), 'start_dt'] = df['end_dt'] - pd.to_timedelta(df['dur_seconds'], unit='S')
    # df.loc[df['end_dt'].isnull() & ~df['start_dt'].isnull() & ~df['dur'].isnull() & ~df['dur_cod'].isnull(), 'end_dt'] = df['start_dt'] + pd.to_timedelta(df['dur_seconds'], unit='S')

    return df
```
