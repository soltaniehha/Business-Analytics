# Data

The data was downloaded from:
https://figshare.com/articles/dataset/CONLIT/21166171/1

We’re working with a dataset called 'CONLIT', which has information on around 2,700 English books from 2001-2021, including their genre, ratings from Goodreads, and more. We're using this data to see what makes some books more popular than others. It contains datasets describing the following:

**CONLIT_META**.**csv**:
Metadata and high level features about each book in the collection listed below.

**CONLIT_POS**.**csv**:
Part-of-speech unigram and bigram frequencies for each book in the collection using Penn Treebank tags.

**CONLIT_UNIGRAM**.**csv**:
Table of 10,000 most frequent word unigrams that appear in at least 18% of all books.

**CONLIT_SUPERSENSE.csv**:
Table of frequencies of 41 word "supersenses" as generated by bookNLP that correspond to WordNet taxonomies spanning both nouns (including plant, animal, food, feeling, and artifact) and verbs (including cognition, communication, motion, etc.).

**CONLIT_WORDCOUNTS**:
Directory of word frequencies for each book in the collection, where each book is represented by a two-column .csv file with the word token and its total frequency in that book.

**CONLIT_WORDCOUNTS_PASSAGES**:
Each book is segmented into 1,000-token passages and token frequencies are provided in the same format as above.

Each row represents a book, each column the book's attributes as described below:

*   **ID**: filename
*   **Category**: refers to fiction or non-fiction (FIC / NON)
*   **Language**: language of text (English)
*   **Genre**: 12 different genre categories
*   **Genre2**: In some cases we note a secondary genre
*   **Pubdate**: Original publication date
*   **Author_Last**: Author last name
* **Author_First**: Author first name
* **Work_Title**: Title of work. In some cases truncated.
* **Translation**: Whether the work was translated from a foreign language.
* **PubHouse**: The publishing company of the original work. Not noted for all works.
* **Prize**: Name of literary prize for which the work was either short list or winner.
* **WinnerShortlist**: Winner or shortlisted.
* **Author_Gender**: Manually annotated for author gender (M/F/O).
* **Author_Nationality**: Author nationality, not noted for all works.
* **token_count**: total number of tokens (includes puncuation)
* **total_characters**: total number of characters. Characters are defined as named entities ("PER") with a proper name associated with them ("PROP") that occur > 5 times.
* **protagonist_concentration**: the number of occurrences of the main character divided by the total number of character occurrences according to the above definition. This is equivalent to a concentration ratio of one (CR1), a standard measure in economics to capture market
concentration by a leading institution.
* **avg_sentence_length**: average length of all sentences in each work.
* **avg_word_length**: average word length of every book.
* **circuitousness:** measure of the difference between the shortest semantic path between all sections of a book and the actual semantic path taken. A proxy for "non-linearity."

* **tuldava_score**: a commonly used measure of reading "difficulty." The formula is:
TD=(Sy/W)x ln(W/St)
where Sy = # syllables, W = # words, St = # sentences

* **event_count**: The number of tokens tagged as a real "event," where a real event is defined as an event that occurs within the diegetic universe of the book (i.e. is not hypothetical). See: Sims, M., Park, J. H., & Bamman, D. (2019, July). Literary event detection. In Proceedings of the 57th Annual Meeting of the Association for Computational Linguistics (pp. 3623-3634).
* **goodreads_avg**: the average rating on Goodreads as of May 23, 2022.
* **total_ratings**: the total number of ratings on Goodreads as of May 23, 2022.
* **goodreads_URL**: URL of each book on Goodreads where ratings information was captured.
* **Probability1P**: The probability of a book being written in the first and third person.

This section examines the potential correlation between Tuldava score (complexity) and Goodreads average rating and tries to make predictions on whether a book's complexity influences how readers rate it.

##Predicting goodreads_avg ratings

To analyze if we can model and predict the values of the average scores of goodreads we are going to preprocess and clean our dataset and then continue with our analysis.\
First we identify the missing values in our dataset and remove the columns that do not provide value for our analysis.

```python
import missingno as msno
msno.matrix(meta_df);
```

```python
# dropping columns with missing values or not of interest for model
drop=['ID','Genre2','Pubdate','Author_Last','Author_First', 'Translation',	'PubHouse',	'Prize',	'WinnerShortlist','Author_Nationality','goodreads_URL','Work_Title']
mdf= meta_df.drop(drop,axis=1)
mdf['Author_Gender'].value_counts()
```

After taking a look at the dataset we can see that it still has 8 values that do not correspond to the gender category and a few missing values. We will remove the 8 values, and the missing ones.

```python
mdf = mdf[mdf['Author_Gender'] != 'o']
```

```python
mdf.dropna(subset=['protagonist_concentration','Author_Gender','volume'],inplace=True) # cleaning missing values
mdf.reset_index()
mdf = pd.get_dummies(mdf, drop_first=True)
mdf.info()
```

Now we can see that the dataset is finally clean and we're ready to train and build our model. We will import the corresponding library from sklearn and then continue with our prediction.

```python
x= mdf.drop('goodreads_avg', axis=1) # preparing our dataset
y= mdf['goodreads_avg']
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3,random_state=833)
Xtrain.shape
```

```python
from sklearn.linear_model import LinearRegression # importing the linear regression model
model= LinearRegression()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

Finally, now that our model is ready, let's see how the predictions are agains the real values

```python
test = Xtest.join(ytest).reset_index()
test.join(pd.Series(y_model, name='predicted')).head(10)
```

```python
fin = test.join(pd.Series(y_model, name='predicted')) # joining our test and predicted values
```

We calculate the mean_absolute_error to compare against the rating avg

```python
from sklearn.metrics import mean_absolute_error # importing the library to compute error
mean_absolute_error(ytest, y_model)
```

We notice that the mean absolute error is 0.2249 which in in rating units, meaning that the mean absolute error of the ratings prediction is 0.225. \
\
Finally we are going to plot the values of both the predicted and real values of goodreads_avg agains the initial tuldava score we had to have an idea of the performance of the model in these 2 dimensions.

```python
g = sns.PairGrid(fin, y_vars=["goodreads_avg","predicted"], x_vars=["tuldava_score"], height=3.5, hue="Category_NON", aspect=1.1) # using pairgrid to visually compare both values.
ax = g.map(plt.scatter, alpha=0.2)

plt.ylim(0, 5)
plt.title( "Goodreads scores and predicted ratings by tuldava score", y=2.05)
g.add_legend();
plt.show()
```

We can see that the predicted values are more condensed then the real ones, but it tends to have a similar distribution across both, fiction and non fiction categories.
